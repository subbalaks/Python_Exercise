```python
               ##################       2.Sentiment Analysis On Movie Description           #####################
```

```python
def classify_category(descList):
    category = []
    sciFi = ['scientist' , 'space station', 'robot']
    action = ['battle' , 'conquer', 'fight']
    for desc in descList:	 
      de = desc.lower()
      # print(str(any(word in de for word in sciFi)))
      if any(word in de for word in sciFi):
          category.append('Sci-Fi')
      elif any(word in de for word in action):
          category.append('Action')
      else:
          category.append('General')
    return category
```

```python
df2 = pd.read_csv("C:\\Users\\ramya\\Desktop\\Python_Assignment\\Exercise_2.csv")
```

```python
#Adding column2 
catList = classify_category(df2['description'].tolist())
df2['Category'] = catList
print(df2)
```
